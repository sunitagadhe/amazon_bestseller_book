# Extracting all the information from both pages of books

```python
#rl=["https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_1?ie=UTF8&pg=1","https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg=2"]
names,authors,prices,ratings=[],[],[],[]
for i in range(1,3):
    url2=f"https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_{i}?ie=UTF8&pg={i}"
    page=requests.get(url2)
    soup=BeautifulSoup(page.content,'html.parser')
    books=soup.find_all(class_="zg-item-immersion")
    for d in books:
        if d.find(class_="a-icon-alt")!=None:
            names.append(d.find(class_="p13n-sc-truncate-desktop-type2").get_text().strip())
            authors.append(d.find(class_="a-row a-size-small").get_text())
            prices.append(d.find(class_="p13n-sc-price").get_text().replace('₹\xa0',''))
            ratings.append(d.find(class_="a-icon-alt").get_text())
        else:
            names.append(d.find(class_="p13n-sc-truncate-desktop-type2").get_text().strip())
            authors.append(d.find(class_="a-row a-size-small").get_text())
            prices.append(d.find(class_="p13n-sc-price").get_text().replace('₹\xa0',''))
            ratings.append(None)     
```

```python
len(names),len(authors),len(prices),len(ratings)
```

```python
import pandas as pd
d={'Names':names,"Authors":authors,'Prices':prices,'Ratings':ratings}
df=pd.DataFrame(d)
df.head(15)
```

```python
df.to_csv('books.csv',index=False)
```
